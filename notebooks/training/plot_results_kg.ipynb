```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.ticker import FormatStrFormatter
```

```python
# base
df_transe_base = pd.read_csv('../../results/kg/metrics_2023-12-16 19 05 05.csv', skiprows=2)
df_transe_base
```

```python
df_transe_1_rel = pd.read_csv('../../results/kg/metrics_2023-12-16 19 07 23.csv', skiprows=2)
df_transe_less_channels = pd.read_csv('../../results/kg/metrics_2023-12-16 19 09 47.csv', skiprows=2)
df_transe_more_channels = pd.read_csv('../../results/kg/metrics_2023-12-16 19 12 08.csv', skiprows=2)
df_complex_1_rel = pd.read_csv('../../results/kg/metrics_2023-12-16 19 14 34.csv', skiprows=2)
df_complex = pd.read_csv('../../results/kg/metrics_2023-12-16 19 16 54.csv', skiprows=2)
df_rotate = pd.read_csv('../../results/kg/metrics_2023-12-16 19 19 15.csv', skiprows=2)
df_rotate_1_rel = pd.read_csv('../../results/kg/metrics_2023-12-16 19 21 36.csv', skiprows=2)
df_distmult_1_rel = pd.read_csv('../../results/kg/metrics_2023-12-16 19 23 36.csv', skiprows=2)
df_distmult = pd.read_csv('../../results/kg/metrics_2023-12-16 19 25 38.csv', skiprows=2)
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4)
fig.set_figwidth(22)
fig.suptitle('Impact of number of hidden channels')
ax1.yaxis.set_major_formatter(FormatStrFormatter('%.1f')), ax2.yaxis.set_major_formatter(FormatStrFormatter('%.3f'))
ax3.yaxis.set_major_formatter(FormatStrFormatter('%.3f')), ax4.yaxis.set_major_formatter(FormatStrFormatter('%.3f'))

ax1.plot(df_transe_base["Epoch"], df_transe_base["Loss"], label='Base number of channels training loss')
ax1.plot(df_transe_more_channels["Epoch"], df_transe_more_channels["Loss"], label='More channels training loss')
ax1.plot(df_transe_less_channels["Epoch"], df_transe_less_channels["Loss"], label='Less channels training loss')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.legend()

ax2.plot(df_transe_base["Epoch"], df_transe_base["Precision"], label='Base number of channels test precision')
ax2.plot(df_transe_more_channels["Epoch"], df_transe_more_channels["Precision"], label='More channels test precision')
ax2.plot(df_transe_less_channels["Epoch"], df_transe_less_channels["Precision"], label='Less channels test precision')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Precision')
ax2.legend()

ax3.plot(df_transe_base["Epoch"], df_transe_base["Recall"], label='Base number of channels test recall')
ax3.plot(df_transe_more_channels["Epoch"], df_transe_more_channels["Recall"], label='More channels test recall')
ax3.plot(df_transe_less_channels["Epoch"], df_transe_less_channels["Recall"], label='Less channels test recall')
ax3.set_xlabel('Epoch')
ax3.set_ylabel('Recall')
ax3.legend()

ax4.plot(df_transe_base["Epoch"], df_transe_base["Hits"], label='Base number of channels test hits ratio')
ax4.plot(df_transe_more_channels["Epoch"], df_transe_more_channels["Hits"], label='More channels test hits ratio')
ax4.plot(df_transe_less_channels["Epoch"], df_transe_less_channels["Hits"], label='Less channels test hits ratio')
ax4.set_xlabel('Epoch')
ax4.set_ylabel('Hits ratio')
ax4.legend()
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4)
fig.set_figwidth(21)
fig.suptitle('Comparision of different models')
ax1.yaxis.set_major_formatter(FormatStrFormatter('%.0f')), ax2.yaxis.set_major_formatter(FormatStrFormatter('%.3f'))
ax3.yaxis.set_major_formatter(FormatStrFormatter('%.3f')), ax4.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))

ax1.plot(df_transe_base["Epoch"], df_transe_base["Loss"], label='TransE training loss')
ax1.plot(df_complex["Epoch"], df_complex["Loss"], label='ComplEx training loss')
ax1.plot(df_rotate["Epoch"], df_rotate["Loss"], label='RotatE training loss')
ax1.plot(df_distmult["Epoch"], df_distmult["Loss"], label='DistMult training loss')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.legend()

ax2.plot(df_transe_base["Epoch"], df_transe_base["Precision"], label='TransE test precision')
ax2.plot(df_complex["Epoch"], df_complex["Precision"], label='ComplEx test precision')
ax2.plot(df_rotate["Epoch"], df_rotate["Precision"], label='RotatE test precision')
ax2.plot(df_distmult["Epoch"], df_distmult["Precision"], label='DistMult test precision')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Precision')
ax2.legend()

ax3.plot(df_transe_base["Epoch"], df_transe_base["Recall"], label='TransE test recall')
ax3.plot(df_complex["Epoch"], df_complex["Recall"], label='ComplEx test recall')
ax3.plot(df_rotate["Epoch"], df_rotate["Recall"], label='RotatE test recall')
ax3.plot(df_distmult["Epoch"], df_distmult["Recall"], label='DistMult test recall')
ax3.set_xlabel('Epoch')
ax3.set_ylabel('Recall')
ax3.legend()

ax4.plot(df_transe_base["Epoch"], df_transe_base["Hits"], label='TransE test hits ratio')
ax4.plot(df_complex["Epoch"], df_complex["Hits"], label='ComplEx test hits ratio')
ax4.plot(df_rotate["Epoch"], df_rotate["Hits"], label='RotatE test hits ratio')
ax4.plot(df_distmult["Epoch"], df_distmult["Hits"], label='DistMult test hits ratio')
ax4.set_xlabel('Epoch')
ax4.set_ylabel('Hits ratio')
ax4.legend()
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4)
fig.set_figwidth(21)
fig.suptitle('Comparision of different models with only user-item relation')
ax1.yaxis.set_major_formatter(FormatStrFormatter('%.1f')), ax2.yaxis.set_major_formatter(FormatStrFormatter('%.3f'))
ax3.yaxis.set_major_formatter(FormatStrFormatter('%.3f')), ax4.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))

ax1.plot(df_transe_1_rel["Epoch"], df_transe_1_rel["Loss"], label='TransE training loss')
ax1.plot(df_complex_1_rel["Epoch"], df_complex_1_rel["Loss"], label='ComplEx training loss')
ax1.plot(df_rotate_1_rel["Epoch"], df_rotate_1_rel["Loss"], label='RotatE training loss')
ax1.plot(df_distmult_1_rel["Epoch"], df_distmult_1_rel["Loss"], label='DistMult training loss')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.legend()

ax2.plot(df_transe_1_rel["Epoch"], df_transe_1_rel["Precision"], label='TransE test precision')
ax2.plot(df_complex_1_rel["Epoch"], df_complex_1_rel["Precision"], label='ComplEx test precision')
ax2.plot(df_rotate_1_rel["Epoch"], df_rotate_1_rel["Precision"], label='RotatE test precision')
ax2.plot(df_distmult_1_rel["Epoch"], df_distmult_1_rel["Precision"], label='DistMult test precision')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Precision')
ax2.legend()

ax3.plot(df_transe_1_rel["Epoch"], df_transe_1_rel["Recall"], label='TransE test recall')
ax3.plot(df_complex_1_rel["Epoch"], df_complex_1_rel["Recall"], label='ComplEx test recall')
ax3.plot(df_rotate_1_rel["Epoch"], df_rotate_1_rel["Recall"], label='RotatE test recall')
ax3.plot(df_distmult_1_rel["Epoch"], df_distmult_1_rel["Recall"], label='DistMult test recall')
ax3.set_xlabel('Epoch')
ax3.set_ylabel('Recall')
ax3.legend()

ax4.plot(df_transe_1_rel["Epoch"], df_transe_1_rel["Hits"], label='TransE test hits ratio')
ax4.plot(df_complex_1_rel["Epoch"], df_complex_1_rel["Hits"], label='ComplEx test hits ratio')
ax4.plot(df_rotate_1_rel["Epoch"], df_rotate_1_rel["Hits"], label='RotatE test hits ratio')
ax4.plot(df_distmult_1_rel["Epoch"], df_distmult_1_rel["Hits"], label='DistMult test hits ratio')
ax4.set_xlabel('Epoch')
ax4.set_ylabel('Hits ratio')
ax4.legend()
```
