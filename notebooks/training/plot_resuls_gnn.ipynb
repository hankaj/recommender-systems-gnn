```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# base
df_lgcn_base = pd.read_csv('../../results/gnn/metrics_2023-11-20 20 40 43.csv', skiprows=2)
df_lgcn_base
```

```python
df_lgcn_xavier = pd.read_csv('../../results/gnn/metrics_2023-11-20 20 51 21.csv', skiprows=2)
df_lgcn_less_layers = pd.read_csv('../../results/gnn/metrics_2023-11-20 21 01 44.csv', skiprows=2)
df_lgcn_more_layers = pd.read_csv('../../results/gnn/metrics_2023-11-26 10 22 22.csv', skiprows=2)
df_lgcn_less_neurons = pd.read_csv('../../results/gnn/metrics_2023-11-20 21 39 04.csv', skiprows=2)
df_lgcn_more_neurons = pd.read_csv('../../results/gnn/metrics_2023-11-20 21 47 55.csv', skiprows=2)
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4)
# fig.set_figheight(10)
fig.set_figwidth(19)
fig.suptitle('Normal init vs Xavier init')

ax1.plot(df_lgcn_base["Epoch"], df_lgcn_base["Loss"], label='Normal init training loss')
ax1.plot(df_lgcn_xavier["Epoch"], df_lgcn_xavier["Loss"], label='Xavier init training loss')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.legend()

ax2.plot(df_lgcn_base["Epoch"], df_lgcn_base["Precision"], label='Normal init test precision')
ax2.plot(df_lgcn_xavier["Epoch"], df_lgcn_xavier["Precision"], label='Xavier init test precision')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Precision')
ax2.legend()

ax3.plot(df_lgcn_base["Epoch"], df_lgcn_base["Recall"], label='Normal init test recall')
ax3.plot(df_lgcn_xavier["Epoch"], df_lgcn_xavier["Recall"], label='Xavier init test recall')
ax3.set_xlabel('Epoch')
ax3.set_ylabel('Recall')
ax3.legend()

ax4.plot(df_lgcn_base["Epoch"], df_lgcn_base["Hits"], label='Normal init test hits ratio')
ax4.plot(df_lgcn_xavier["Epoch"], df_lgcn_xavier["Hits"], label='Xavier init test hits ratio')
ax4.set_xlabel('Epoch')
ax4.set_ylabel('Hits ratio')
ax4.legend()
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4)
fig.set_figwidth(19)
fig.suptitle('Impact of number of layers')

ax1.plot(df_lgcn_base["Epoch"], df_lgcn_base["Loss"], label='Base number of layers training loss')
ax1.plot(df_lgcn_more_layers["Epoch"], df_lgcn_more_layers["Loss"], label='More layers training loss')
ax1.plot(df_lgcn_less_layers["Epoch"], df_lgcn_less_layers["Loss"], label='Less layers training loss')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.legend()

ax2.plot(df_lgcn_base["Epoch"], df_lgcn_base["Precision"], label='Base number of layers test precision')
ax2.plot(df_lgcn_more_layers["Epoch"], df_lgcn_more_layers["Precision"], label='More layers test precision')
ax2.plot(df_lgcn_less_layers["Epoch"], df_lgcn_less_layers["Precision"], label='Less layers test precision')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Precision')
ax2.legend()

ax3.plot(df_lgcn_base["Epoch"], df_lgcn_base["Recall"], label='Base number of layers test recall')
ax3.plot(df_lgcn_more_layers["Epoch"], df_lgcn_more_layers["Recall"], label='More layers test recall')
ax3.plot(df_lgcn_less_layers["Epoch"], df_lgcn_less_layers["Recall"], label='Less layers test recall')
ax3.set_xlabel('Epoch')
ax3.set_ylabel('Recall')
ax3.legend()

ax4.plot(df_lgcn_base["Epoch"], df_lgcn_base["Hits"], label='More layers test hits ratio')
ax4.plot(df_lgcn_more_layers["Epoch"], df_lgcn_more_layers["Hits"], label='More layers test hits ratio')
ax4.plot(df_lgcn_less_layers["Epoch"], df_lgcn_less_layers["Hits"], label='Less layers test hits ratio')
ax4.set_xlabel('Epoch')
ax4.set_ylabel('Hits ratio')
ax4.legend()
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4)
fig.set_figwidth(19)
fig.suptitle('Impact of number of neurons')

ax1.plot(df_lgcn_base["Epoch"], df_lgcn_base["Loss"], label='Base number of neurons training loss')
ax1.plot(df_lgcn_more_neurons["Epoch"], df_lgcn_more_neurons["Loss"], label='More neurons training loss')
ax1.plot(df_lgcn_less_neurons["Epoch"], df_lgcn_less_neurons["Loss"], label='Less neurons training loss')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.legend()

ax2.plot(df_lgcn_base["Epoch"], df_lgcn_base["Precision"], label='Base number of neurons test precision')
ax2.plot(df_lgcn_more_neurons["Epoch"], df_lgcn_more_neurons["Precision"], label='More neurons test precision')
ax2.plot(df_lgcn_less_neurons["Epoch"], df_lgcn_less_neurons["Precision"], label='Less neurons test precision')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Precision')
ax2.legend()

ax3.plot(df_lgcn_base["Epoch"], df_lgcn_base["Recall"], label='Base number of neurons test recall')
ax3.plot(df_lgcn_more_neurons["Epoch"], df_lgcn_more_neurons["Recall"], label='More neurons test recall')
ax3.plot(df_lgcn_less_neurons["Epoch"], df_lgcn_less_neurons["Recall"], label='Less neurons test recall')
ax3.set_xlabel('Epoch')
ax3.set_ylabel('Recall')
ax3.legend()

ax4.plot(df_lgcn_base["Epoch"], df_lgcn_base["Hits"], label='More layers test hits ratio')
ax4.plot(df_lgcn_more_neurons["Epoch"], df_lgcn_more_neurons["Hits"], label='More neurons test hits ratio')
ax4.plot(df_lgcn_less_neurons["Epoch"], df_lgcn_less_neurons["Hits"], label='Less neurons test hits ratio')
ax4.set_xlabel('Epoch')
ax4.set_ylabel('Hits ratio')
ax4.legend()
```
